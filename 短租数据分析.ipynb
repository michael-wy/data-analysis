```python
places = {'朝阳区 / Chaoyang':'Chaoyang',
         '密云县 / Miyun':'Miyun',
         '东城区':'Dongcheng',
         '西城区':'Xicheng',
         '海淀区':'Haidian',
         '顺义区 / Shunyi':'Shunyi',
         '房山区':'Fangshan',
         '怀柔区 / Huairou':'Huairou',
         '昌平区':'Changpin',
         '通州区 / Tongzhou':'Tongzhou',
         '丰台区 / Fengtai':'Fengtai',
         '大兴区 / Daxing':'Daxing',
         '延庆县 / Yanqing':'Yanqing',
         '石景山区':'Shijingshan',
         '门头沟区 / Mentougou':'Mentougou',
         '平谷区 / Pinggu':'Pinggu'}
print(type(places))
data['place'] = data['neighbourhood'].map(places)
data['place'].value_counts()
#value_counts()是一种查看表格某列中有多少个不同值的快捷方法，并计算每个不同值有在该列中有多少重复值。
```

```python
bins = np.array([0,100,200,300,400,500,600,700,800,900,1000,10000,100000])
data.loc[data['price'].idxmax()]
```

```python
data.drop(index=[data['price'].idxmax()],inplace = True)
data.loc[data['price'].idxmax()]
```

```python
labels = pd.cut(data['price'],bins)
print(pd.value_counts(labels))
```

```python
sum_price = pd.value_counts(labels)
sum_price.plot(kind = 'bar')
```
