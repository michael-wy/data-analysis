```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
speed_dist=pd.read_csv(r'H:\62882【数据分析从入门到进阶】配套数据\【数据分析从入门到进阶】配套数据\第1章\cars.csv')
speed_dist.head()
```

```python
#绘制散点图，并添加拟合线
sns.lmplot(x='speed',#指定X轴变量
          y='dist',#指定Y轴变量
          data=speed_dist,#指定绘图数据集
          legend_out=False,#将图例呈现在图框内
          truncate=False#根据实际的数据范围，对拟合线做截断操作
          )
#显示图形
plt.show()
```
