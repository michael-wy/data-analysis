```python
apply()与applymap()函数
我们在python基础知识部分讲过一个python的高级特性map()函数，map()函数是对一个序列中的所有元素执行相同的操作。
在DataFrame中与map函数类似的函数有两个，一个是apply()函数，另一个是applymap()函数,函数apply()和applymap()都需要与匿名函数lamba结合使用。
apply()函数主要用于对DataFrame中的某一column或row中元素执行相同的函数操作。
```

```python
a={"C1":[1,4,7],"C2":[2,5,8],"C3":[3,6,9]}
df=pd.DataFrame(a)
df
```

```python
# 对C1列汇中的每个元素加1
df['C1'].apply(lambda x :x+1)
```

```python
# applymap()函数用对于DataFrame中的每一个元素执行相同的函数操作
df.applymap(lambda x :x+1)
```
